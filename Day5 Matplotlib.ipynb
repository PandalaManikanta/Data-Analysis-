## matplotlib is a python library for visualizastion part

- lineplot
- scatterplot
- barplot
- piechart
- boxplot
- histogram

```python
import matplotlib.pyplot as plt
```

## r-- is used to change the color of line(r=red,g=green)

## * is used to mark the matched point with star

```python
x=[1.1,1.2,2.1,2.2]
y=[8.8,8.2,8.7,7.9]
plt.plot(x,y,'r*--')
plt.show()
```

```python
x=[1.1,1.2,2.1,2.2,3.1]
y=[8.8,8.2,8.7,7.9,9.2]
plt.plot(x,y,'m*--')
plt.show()
```

```python
help(plt)
```

## ^ for traingle marks

```python
plt.plot(x,y,"^")
```

## go is used for circle marks

```python
plt.plot(x,y,"go")
```

## grid view

```python
plt.plot(x,y,"go")
plt.plot(y,x,"c*")## c* is a skycolor star
plt.grid()
plt.show()
```

```python
plt.plot(x,y,"go",label="xyvalues")
plt.plot(y,x,"c*",label="yxvalues")## c* is a skycolor star
plt.legend()
plt.grid()
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

```python
plt.plot(x,y,"go",label="xyvalues")
plt.plot(y,x,"c*",label="yxvalues")## c* is a skycolor star
plt.legend(loc="upper left")
plt.grid()
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

```python
plt.plot(x,y,"go",label="xyvalues")
plt.plot(y,x,"c*",label="yxvalues")## c* is a skycolor star
plt.legend(loc="lower right")
plt.grid()
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

```python
x=[1,3,4,5]
y=[3,4,5,6]
plt.scatter(x,x,color="red")
plt.scatter(x,y,color='g')
plt.grid()
```

## Bargraph

```python
x=[60,70,40,80,30,50]
y=[50,40,60,90,60,30]
plt.grid()
plt.bar(x,y,color='y',label="Marks")# k=black,r=red,sb=skyblue
plt.legend(loc="upper right")
```

## bar is for vertical graph

```python
x=[60,70,40,80,30,50]
y=[50,40,60,90,60,30]
plt.grid()
plt.bar(x,y,color='y',width=5,label="Marks")# k=black,r=red,sb=skyblue
plt.legend()
```

## barh is for horizontal graph

```python
plt.barh(x,y,color='y',height=5)
```

```python
x=["mani","anil","raj","abhi","raghu"]
y=[84,55,37,66,57]
plt.bar(x,y,color="g",label="Marks")
plt.legend()
plt.title("Python Marks")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
plt.savefig("pythonmarks.jpg")
```

```python
x=["mani","anil","raj","abhi","raghu"]
y=[84,55,37,66,57]
plt.barh(x,y,color="g",label="Marks")
plt.legend()
plt.title("Python Marks")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
plt.savefig("pythonmarks.jpg")
```

## Histogram

```python
x=np.array([1,3,4,5,6,0])
plt.hist(x,bins=5)
```

```python
y=np.random.randint(1,100,5)
plt.hist(y,bins=10)
```

##  Pie graphs

```python
x=["cse","it","ece","eee"]
y=[240,60,240,60]
plt.pie(y,labels=x)
```

```python
plt.pie(y,labels=x,explode=[0.2,0.2,0,0])# 0.2 is distance  to seperate the pie graph
```

```python
plt.pie(y,labels=x,explode=[2,0.2,0,0.5])
```

```python
plt.pie(y,labels=x,explode=[0.5,0.5,0.5,0.5])
```

## box plot

```python
x=[60,50,40,30]# yellow line is median
plt.boxplot(x)# low line is minimum and high line is maximum
```

```python
x=[60,50,40,30]
y=[24,60,27,6]
z=[44,50,66,99]
data=[x,y,z]
plt.boxplot(data)
```

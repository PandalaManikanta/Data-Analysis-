## matplotlib is a python library for visualizastion part

- lineplot
- scatterplot
- barplot
- piechart
- boxplot
- histogram

```python
import matplotlib.pyplot as plt
```

## r-- is used to change the color of line(r=red,g=green)

## * is used to mark the matched point with star

```python
x=[1.1,1.2,2.1,2.2]
y=[8.8,8.2,8.7,7.9]
plt.plot(x,y,'r*--')
plt.show()
```

```python
x=[1.1,1.2,2.1,2.2,3.1]
y=[8.8,8.2,8.7,7.9,9.2]
plt.plot(x,y,'m*--')
plt.show()
```

```python
help(plt)
```

## ^ for traingle marks

```python
plt.plot(x,y,"^")
```

## go is used for circle marks

```python
plt.plot(x,y,"go")
```

## grid view

```python
plt.plot(x,y,"go")
plt.plot(y,x,"c*")## c* is a skycolor star
plt.grid()
plt.show()
```

```python
plt.plot(x,y,"go",label="xyvalues")
plt.plot(y,x,"c*",label="yxvalues")## c* is a skycolor star
plt.legend()
plt.grid()
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

```python
plt.plot(x,y,"go",label="xyvalues")
plt.plot(y,x,"c*",label="yxvalues")## c* is a skycolor star
plt.legend("right")
plt.grid()
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

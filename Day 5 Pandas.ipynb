## merging in a dataframe

```python
df1=pd.DataFrame({"city":["nrt","gnt",'vij'],"temp":[32,31,30]})
df1
```

```python
df2=pd.DataFrame({"city":["nrt","gnt","vij"],"humb":[32,54,34]})
df2
```

```python
df3=pd.merge(df1,df2)
df3
```

```python
df3=pd.merge(df1,df2,on="city")
df3
```

## only same name columns will be merged remaining will not merge

```python
t=pd.DataFrame({"city":["raj","gnt","vij"],"humb":[32,54,34]})
```

```python
e=pd.DataFrame({"city":["nrt","gnt",'vij'],"temp":[32,31,30]})
```

```python
d=pd.merge(t,e)
d
```

## inner =same values will print

```python
d=pd.merge(t,e,on="city",how="inner")
d
```

## outer=different values will  print

```python
d=pd.merge(t,e,on="city",how="outer")
d
```

```python
d=pd.merge(t,e,on="city",how="right")
d
```

## right will print right values
## left will print the left values

```python
d=pd.merge(t,e,on="city",how="left")
d
```

## indicator will represents the merged status

```python




d=pd.merge(t,e,on="city",how="right",indicator=True)
d
```
